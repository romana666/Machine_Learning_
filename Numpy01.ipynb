## Numpy-Numerical python. It keeps homogeneous data
## 1. ndarray ( N dimensional array) and its Attributes


## 1.1.ndarray

```python
import numpy as np



# 1d array creating and checking

arr1 = np.array([1,2,3,4,5])

print(type(arr1))
print(arr1.ndim)      #Attribute 1. ndim



# 2 d array

arr2 = np.array([[1,2,3,4],    #row1
                [4,5,6,7]])   #row2

print(arr2.ndim)



# 3 d array  - Multiple 2d array in  1d array

arr3 = np.array([
          [[ 1,2,4],           #1st floor
           [ 3,4,5]],

           [[2,4,5],            #2nd floor
            [3,4,5]],

            [[2,4,5],            #3rd floor
            [3,4,9]]

])

print(arr3.ndim)
```

## 2. Array Attributes

```python



#1.ndim
print(arr1.ndim)
print(arr2.ndim)
print(arr3.ndim)


#2.shape
print(arr1.shape)
print(arr2.shape)
print(arr3.shape)

#3.data type
print(arr1.dtype)
print(arr2.dtype)
print(arr3.dtype)

#4.size
print(arr1.size)
print(arr2.size)
print(arr3.size)
```

# 3. nd array data type

```python
arr = np.array([1,2,3,4])
print(arr.dtype)


# upcasting into float
arr = np.array([1.3,2.5,3,5.5])
print(arr.dtype)

#upcasting into string
arr = np.array ([1,2,3.2,"hello"])
print(arr.dtype)
```

```python
#Selecting datatype

arr = np.array([1,2,300],dtype = np.int32)
print(arr.dtype)

```

```python
arr =arr.astype(np.int16)
print(arr.dtype)
```

## 4.nd array creation from existing data

```python
#list
#1.from homogeneous list
lst = [10,20,30,40.0]

arr = np.array(lst,dtype = np.int32)

print(arr)
print(type(arr))
print(arr.dtype)

#2.from non homogeneous list

mixed_list = [10,True,'hello']  #But not possible to keep another list i9n one list,for that execution i have to keep homogeneous

arr = np.array(mixed_list)

print(arr)
print(type(arr))
print(arr.dtype)

#3.from 2d list

matrix = [[1,2,3,4],
          [4,5,6,7]]
arr = np.array(matrix)

print(arr)
print(type(arr))
print(arr.dtype)
print(arr.shape)
print(arr.ndim)


```

```python
#tuple

tpl = (1,2,3)
arr = np.array(tpl,dtype = np.int16)
arr = arr.astype(np.int32)

print(arr)
print(type(arr))
print(arr.dtype)
print(arr.shape)
print(arr.ndim)
```

```python
#set

st = {1,2,3}
arr = np.array([*st])  #was facing issue on arr= np.array(list(st))
print(arr.dtype)
#arr = arr.astype(np.int32)

print(arr)
print(type(arr))
print(arr.shape)
print(arr.ndim)
```

```python
#dictionary

dc ={'a':10,'b':20,'c':30}

keys = dc.keys()
values = dc.values()
items = dc.items()

print(keys,values,items)


#keys to ndarray

arr = np.array([*keys])  #because of string type change not executed
print(arr)
print(type(arr))
print(arr.dtype)
print(arr.shape)
print(arr.ndim)
print(arr.size)


#values to ndarray

arr = np.array([*values],dtype = np.int32)
print(type(arr))
print(arr.dtype)
print(arr.shape)
print(arr.ndim)
print(arr.size)

#items to ndarray

arr = np.array([*items])
print(arr)
print(type(arr))
print(arr.dtype)
print(arr.shape)
print(arr.ndim)
print(arr.size)


```

##  5.ndarray generation with random values

```python
#np.random.rand(shape)

arr = np.random.rand(2,3)  #0->1 diistributed

print(arr)
print(type(arr))
print(arr.dtype)
print(arr.shape)
print(arr.ndim)
print(arr.size)


#np.random.randint(strat,end,shape)

arr = np.random.randint(1,10,(2,2,2))

print(arr)
print(type(arr))
print(arr.dtype)
print(arr.shape)
print(arr.ndim)
print(arr.size)

#np.random.uniform(strat,end,shape) -float

arr = np.random.uniform(50,100,(3,4))

print(arr)
print(type(arr))
print(arr.dtype)
print(arr.shape)
print(arr.ndim)
print(arr.size)
```

## 6.ndarray creation from range

```python
#np.arange(first,end,step)  -> similliar to range function

arr = np.arange(1,10,1)
mat = arr.reshape(3,3)       #->reshape
print(arr)
print(mat)
print(type(arr))
print(arr.dtype)
print(arr.shape)
print(arr.ndim)
print(arr.size)
```

```python
#linspace
#np.linspace(start , end , how many)

arr = np.linspace(1,10,12)

print(arr)
print(type(arr))
print(arr.dtype)
print(arr.shape)
print(arr.ndim)
print(arr.size)

```

```python
#np.logspace(start, end (included),koita)
arr = np.logspace(0,4,5)
arr1 = np.logspace(0,4,6,base = 2)
print(arr)
print(arr1)
```
