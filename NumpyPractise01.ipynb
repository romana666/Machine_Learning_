```python
import numpy as np
import pandas as pd

np.random.seed(24)

num_employees = 50  # you can increase this if needed

# Employee data
emp_ids = np.arange(101, 101 + num_employees)
emp_ages = np.random.randint(20, 55, num_employees)
emp_income = np.random.randint(35000, 95000, num_employees)

# Fixed set of roles (repetition allowed)
roles_list = [
    "HR Manager",
    "Software Developer",
    "Accountant",
    "Marketing Executive",
    "Sales Executive",
    "Backend Developer",
    "HR Executive",
    "Financial Analyst",
    "Digital Marketer",
    "Operations Manager"
]

emp_roles = np.random.choice(roles_list, num_employees)

# Create DataFrame
data = pd.DataFrame({
    "emp_id": emp_ids,
    "emp_age": emp_ages,
    "emp_salary": emp_income,
    "role": emp_roles
})

# Save dataset
data.to_csv("staff_data.csv", index=False)
```

```python
#1.
#emp_ages
print("Employee Ages's Data type : ",type(emp_ages))
print("Employee Ages's Dimension : ",emp_ages.ndim)
print("Employee Ages's shape : ", emp_ages.shape)
print("Employee  Ages's total number of element : ", emp_ages.shape)


print("")

#emp_income
print("Employee Incomes's Data type : ",type(emp_income))
print("Employee Incomes's Dimension : ",emp_income.ndim)
print("Employee Incomes's shape : ", emp_income.shape)
print("Employee  Incomes's total number of element : ", emp_income.shape)
```

```python
#2
#Find the maximum and minimum salary
max_sal = np.max(emp_income)
min_sal = np.min(emp_income)

#Calculate the average salary
avg_sal = np.mean(emp_income)

#Calculate the average age
avg_age = np.mean(emp_ages)

print("Maximum Salary : ", max_sal)
print("Minimum Salary : ", min_sal)
print("Average Salary : ", avg_sal)
print("Average Age : ", avg_age)
```

```python
#3.
filtered_employees = data[
                     (data["emp_age"] >= 25) &
                     (data["emp_age"] <= 40) &
                     (data["emp_salary"] > 50000)]

# Filtered values
print("Filtered Employees:")
print(filtered_employees)

# Count satisfy employees
count = filtered_employees.shape[0]

print("")

print("Number of employees satisfying the condition:", count)
```

```python
#4.
original_ages = emp_ages
avg_age = np.mean(original_ages)

#Updated array without modifying original
updated_ages = np.where(original_ages < avg_age,
                        original_ages + 4,
                        original_ages + 2)


print("Original Age Array : ",original_ages)
print("")
print("Updated Age Array : ",updated_ages)
print("")
print("Average Age Used for Condition : ", avg_age)
```

```python
#5.
roles = data["role"].to_numpy()
salary = data["emp_salary"].to_numpy()

old_total = np.sum(salary)

new_salary = np.where(np.isin(roles, ["Software Developer", "Backend Developer"]),
                      salary * 1.08,
                      salary)

new_total = np.sum(new_salary)

print("Old Total:", old_total)
print("New Total:", new_total)
print("Increase:", new_total - old_total)
```
