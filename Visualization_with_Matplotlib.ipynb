## 3.Categorical (One - Variable)

```python
data = {
    'Student': [ 'B', 'C', 'D', 'E', 'F', 'G', 'H'],
    'Gender': [ 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male'],
    'StudyHours': [ 2, 5, 3, 7, 4, 8, 1],
    'Attendance': [ 60, 88, 70, 95, 80, 98, 55],
    'Grade': [ 'Fail', 'Pass', 'Fail', 'Pass', 'Pass', 'Pass', 'Fail']
}

df = pd.DataFrame(data)
```

```python
df
```

```python
#Bar Chart -- Count/Frequency  (cat vs cat)

#gender wise
gender_group = df.groupby("Gender").size()
gender_group.index
gender_group.values
```

```python
plt.bar(gender_group.index,gender_group.values,color=["purple","blue"],edgecolor="black")

plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Bar Chart of Gender among students')
```

```python
#Pie Chart -- Percentage/Proportion

#Grade wise
grade = df.groupby("Grade").size()
plt.pie(grade,labels = grade.index,autopct = "%1.1f%%",shadow =True)
```

## 4. Numerical VS Categorical

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "Department": ["CSE", "EEE", "BBA", "CSE", "EEE", "BBA", "CSE", "EEE", "BBA"],
    "Salary": [50000, 45000, 40000, 55000, 47000, 42000, 60000, 49000, 43000]
}

df = pd.DataFrame(data)

print(df)
```

```python
#Bar plot
avg_salary = df.groupby("Department")["Salary"].mean()

plt.bar(avg_salary.index, avg_salary.values)

plt.title("Average Salary by Department")
plt.xlabel("Department")
plt.ylabel("Average Salary")

plt.show()
```

```python
#Box Plot
cse = df[df["Department"] == "CSE"]["Salary"]
eee = df[df["Department"] == "EEE"]["Salary"]
bba = df[df["Department"] == "BBA"]["Salary"]

plt.boxplot([cse, eee, bba],labels=["CSE", "EEE", "BBA"])

plt.title("Salary Distribution by Department")
plt.xlabel("Department")
plt.ylabel("Salary")
```
