```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
data = {
    "Age": [20, 22, 21, 23, 24, 25, 22, 21, 23, 24],
    "Salary": [30000, 35000, 32000, 40000, 45000, 50000, 37000, 33000, 42000, 47000],
    "Department": ["CSE", "EEE", "CSE", "BBA", "CSE", "EEE", "BBA", "CSE", "EEE", "BBA"]
}

df = pd.DataFrame(data)
df
```

##1.Numerical (One Variable)

```python
#Histogram (Also num vs num)
sns.histplot(data=df, x="Salary")
plt.show()
```

```python
#KDE plot (Also num vs num)
sns.kdeplot(data=df, x="Salary")
plt.show()
```

```python
#Line Plot
sns.lineplot(data=df,x=df.index, y="Salary")
plt.show()
```

```python
#Scatter plot
sns.scatterplot(data=df, x="Salary",y=df.index,hue = "Department")
plt.show()
```

## 2.Numerical VS Numerical

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {
    "Age": [20, 22, 23, 25, 27, 29, 30, 32, 35, 37],
    "Height": [160, 162, 165, 167, 170, 172, 174, 176, 178, 180],
    "Weight": [52, 55, 58, 60, 64, 67, 70, 72, 75, 78],
    "Salary": [25000, 28000, 30000, 35000, 40000, 45000, 50000, 55000, 62000, 70000],
    "Gender":["F","M",'F','F','F','M','M','F','M','M']
}

df = pd.DataFrame(data)

print(df)
```

```python
#Line Plot
sns.lineplot(data = df,x="Age",y="Salary",hue ='Gender')
```

```python
# Scatterplot
sns.scatterplot(data = df, x ="Height",y = "Weight",hue='Gender')
plt.show()
```

```python
#Regression Plot
sns.regplot(data = df,x="Height",y="Weight")
plt.show()
```

```python
#Joint plot
sns.jointplot(data =df,x="Height",y = "Weight",hue = 'Gender')
```

```python
#KDE
sns.kdeplot(data =df,x= "Height",y="Weight")
```

```python
sns.kdeplot(
    data=df,
    x="Height",
    y="Weight",
    fill=True
)

plt.show()
```

```python
#Pairplot
sns.pairplot(df)
```

```python
#Heatmap
sns.heatmap(df.corr())
plt.show()
```

## 3.Categorical Vs Categorical


```python


data = {
    "Department": [
        "CSE","CSE","EEE","EEE","BBA",
        "BBA","CSE","EEE","BBA","CSE"
    ],
    "Gender": [
        "Male","Female","Male","Female","Male",
        "Female","Male","Female","Male","Female"
    ]
}

df = pd.DataFrame(data)

print(df)
```

```python
#Countplot
sns.countplot(data=df,x="Department",hue="Gender")

plt.show()
```

```python
#Heatmap - using crosstab
table = pd.crosstab(df["Department"],df["Gender"])

sns.heatmap(table,annot=True,cmap="Blues")

plt.show()
```

## 4.Numerical vs Categorical

```python


data = {
    "Student": ["A", "B", "C", "D", "E", "F"],
    "Department": ["CSE", "EEE", "CSE", "BBA", "EEE", "BBA"],
    "Salary": [50000, 45000, 60000, 40000, 55000, 42000]
}

df = pd.DataFrame(data)
```

```python
#Barplot
sns.barplot(
    data=df,
    x="Department",
    y="Salary"
)

plt.title("Average Salary by Department")
plt.show()
```

```python
sns.boxplot(data=df,x="Department",y="Salary")

plt.title("Salary Distribution by Department")
plt.show()
```
